#1. Análisis de facturación



```python
def suma_x_tienda(td):
    return td['Precio'].sum()

suma_t1 = suma_x_tienda(tienda)
suma_t2 = suma_x_tienda(tienda2)
suma_t3 = suma_x_tienda(tienda3)
suma_t4 = suma_x_tienda(tienda4)

print('Ingreso total tienda 1: ', int(suma_t1))
print('Ingreso total tienda 2: ', int(suma_t2))
print('Ingreso total tienda 3: ', int(suma_t3))
print('Ingreso total tienda 4: ', int(suma_t4))
```

```python

import matplotlib.pyplot as plt


ingresos = [suma_t1, suma_t2, suma_t3, suma_t4]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

plt.figure(figsize=(9, 6))
barras = plt.bar(tiendas, ingresos, color='#4CAF50', edgecolor='black')

plt.title('Ingresos por Tienda', fontsize=16, fontweight='bold')
plt.xlabel('Tiendas', fontsize=12)
plt.ylabel('Ingresos ($)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)

for barra, valor in zip(barras, ingresos):
    altura = barra.get_height()
    plt.text(barra.get_x() + barra.get_width() / 2,
             altura + max(ingresos) * 0.02,
             f"${valor:,.0f}",
             ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
def conteo_categorias(df, top_n=5):
    conteo = df['Categoría del Producto'].value_counts().head(top_n)
    print(conteo)
    return conteo

tiendas = [tienda, tienda2, tienda3, tienda4]

for i, t in enumerate(tiendas, start=1):
    print(f"Categorías más vendidas en Tienda {i}:")
    conteo_categorias(t, top_n=5)
    print()
```

```python
import matplotlib.pyplot as plt

def graficar_categorias_mas_vendidas(df, tienda_num, top_n=5):
    conteo = df['Categoría del Producto'].value_counts().head(top_n)
    categorias = conteo.index.tolist()
    cantidades = conteo.values.tolist()

    plt.figure(figsize=(8,5))
    barras = plt.bar(categorias, cantidades, color='skyblue', alpha=0.7)
    plt.title(f'Top {top_n} Categorías más vendidas - Tienda {tienda_num}')
    plt.xlabel('Categorías')
    plt.ylabel('Cantidad de productos vendidos')
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.6)

    for barra, valor in zip(barras, cantidades):
        altura = barra.get_height()
        plt.text(barra.get_x() + barra.get_width()/2, altura + max(cantidades)*0.02,
                 f'{valor}', ha='center', va='bottom', fontsize=10, fontweight='bold')

    plt.tight_layout()
    plt.show()

tiendas = [tienda, tienda2, tienda3, tienda4]

for i, t in enumerate(tiendas, start=1):
    graficar_categorias_mas_vendidas(t, i, top_n=5)
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
from adjustText import adjust_text
import folium
from folium.plugins import HeatMap
import pandas as pd

df = pd.concat([tienda.assign(Tienda='Tienda 1'),
                tienda2.assign(Tienda='Tienda 2'),
                tienda3.assign(Tienda='Tienda 3'),
                tienda4.assign(Tienda='Tienda 4')])

df['Ingreso'] = df['Precio'] * df['Cantidad de cuotas']

ingresos_por_ubicacion = df.groupby(['lat', 'lon', 'Lugar de Compra'])['Ingreso'].sum().reset_index()

plt.figure(figsize=(12,8))
scatter = plt.scatter(
    ingresos_por_ubicacion['lon'],
    ingresos_por_ubicacion['lat'],
    s=ingresos_por_ubicacion['Ingreso'] / ingresos_por_ubicacion['Ingreso'].max() * 500,
    c=ingresos_por_ubicacion['Ingreso'],
    cmap='RdYlGn',
    alpha=0.7
)

cbar = plt.colorbar(scatter, label='Ingreso Total')
cbar.ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{int(x):,}'))

plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.title('Distribución geográfica de ingresos por ubicación')

texts = []
for i, row in ingresos_por_ubicacion.iterrows():
    texts.append(plt.text(row['lon'], row['lat'], row['Lugar de Compra'], fontsize=8, ha='right'))

adjust_text(texts, arrowprops=dict(arrowstyle='-', color='gray'))

plt.show()

mapa = folium.Map(location=[df['lat'].mean(), df['lon'].mean()], zoom_start=6)

max_ingreso = ingresos_por_ubicacion['Ingreso'].max()
heat_data = [[row['lat'], row['lon'], row['Ingreso'] / max_ingreso] for index, row in ingresos_por_ubicacion.iterrows()]

HeatMap(heat_data, radius=15).add_to(mapa)

mapa.save('mapa_ventas.html')


```

## Costo de Envío Promedio

El costo de envío promedio está directamente relacionado con el volumen de ventas y, por ende, con la facturación de cada tienda. Consistentemente, la **Tienda 1, con la mayor facturación, incurre en el mayor costo de envío promedio**, mientras que la **Tienda 4, con la menor facturación, presenta el costo de envío promedio más bajo**.

Es relevante notar que, a pesar de su menor facturación, la **Tienda 4 mantiene porcentajes de venta significativos en categorías clave como 'Electrónicos' y 'Muebles', y a la vez presenta un costo de envío promedio menor**. Esto podría indicar una gestión logística más eficiente o un menor costo por envío unitario en comparación con las otras tiendas.

### Puntos Clave:

* El costo de envío promedio refleja el volumen de ventas de cada tienda.
* La Tienda 4 podría tener prácticas de envío más eficientes o costos unitarios menores.
* Analizar la relación entre el costo de envío y el valor de los productos enviados por tienda podría ofrecer información adicional.

## Conclusión y Recomendaciones Mejoradas

Considerando el análisis integral de los datos, la **baja facturación de las Tiendas 2, 3 y 4** podría ser atribuible a diversos factores, incluyendo la ubicación geográfica, la antigüedad de la tienda, estrategias de marketing, gestión del inventario o la eficiencia operativa. La rentabilidad real de cada tienda solo podría determinarse al considerar los costos operativos asociados.

La recomendación inicial de vender la Tienda 4, basada en su menor facturación, debe ser reconsiderada a la luz de otros factores positivos identificados:

* **Mejor calificación promedio de los clientes:** Indica una mayor satisfacción del cliente en comparación con la Tienda 1.
* **Menor costo de envío promedio:** Sugiere una posible eficiencia en la logística o menores costos asociados al envío.
* **Fuerza relativa en la venta de 'Juguetes':** Podría representar un nicho de mercado valioso.

4.  **Evaluación del Potencial del Nicho de 'Juguetes' en la Tienda 4:** Considerar si este nicho puede ser explotado o replicado en otras tiendas.
5.  **Análisis de la Estrategia de Productos:** Evaluar si la oferta homogénea de productos en todas las tiendas es óptima o si se podrían introducir variaciones basadas en la demanda local o en las fortalezas de cada tienda (como los 'Juguetes' en la Tienda 4).
6.  **Auditoría Operacional y de Gestión:** Identificar posibles ineficiencias o áreas de mejora en la gestión de cada tienda que puedan estar afectando su rendimiento.
